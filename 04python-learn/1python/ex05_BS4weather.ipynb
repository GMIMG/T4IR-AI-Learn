```python
import requests
from bs4 import BeautifulSoup
response = requests.get('https://web.kma.go.kr/weather/observation/currentweather.jsp')
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
table=soup.find(class_="table_develop3")
```

```python
table
```

```python
lst=[]
for datas in table.find('tbody').find_all('tr'):
    data = datas.find_all('td')
    str = data[0].text + " : "+ data[5].text + ", " + data[10].text
    lst.append([data[0].text, data[5].text, data[10].text])
    print(str)
lst[0][2]
```
