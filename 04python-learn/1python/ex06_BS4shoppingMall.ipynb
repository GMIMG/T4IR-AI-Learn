```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://sonyunara.com/shop/list.php?cate=0104"
```

```python
response = requests.get(url)
```

```python
response.encoding = "utf-8"
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text)
```

```python
ul=soup.find(class_='products-list')
```

```python
li_lst=ul.find_all(class_='item')
```

```python
data = []
for li in li_lst:
    s = li.find(class_='subject').text
    p = li.find(class_='price').find('strong').text.replace(',','')
    data.append([s,p])
```

```python
data
```

```python
with open('product.csv','w') as file:
    file.write('product_name,price\n')
    for i in data:
        file.write('{},{}\n'.format(i[0],i[1]))
```
