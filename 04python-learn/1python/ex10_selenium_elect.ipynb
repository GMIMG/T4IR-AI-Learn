```python
from selenium import webdriver
import pandas as pd
import numpy as np
import time
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get("http://info.nec.go.kr/main/showDocument.xhtml?electionId=0000000000&topMenuId=VC&secondMenuId=VCCP09")
time.sleep(1)
```

```python
driver.find_element_by_id('electionType1').click()
time.sleep(1)
```

```python
driver.find_element_by_xpath('//*[@id="electionName"]/option[@value='+str(20170509)+']').click()
time.sleep(1)
```

```python
driver.find_element_by_xpath('//*[@id="electionCode"]/option[@value='+str(1)+']').click()
time.sleep(1)
```

```python
driver.find_element_by_xpath('//*[@id="cityCode"]/option[@value='+str(0)+']').click()
time.sleep(1)
```

```python
driver.find_element_by_id('searchBtn').click()
time.sleep(1)
```

```python
table = driver.find_element_by_id('table01')
people = table.find_elements_by_class_name('alignC')
```

```python
data= []
tb = driver.find_element_by_tag_name('tbody')
tr_lst = tb.find_elements_by_tag_name('tr')
for tr in tr_lst:
    td_lst = tr.find_elements_by_tag_name('td')
    temp = []
    for td in td_lst:
        temp.append(td.text.replace(',','').replace('\n',''))
    data.append(temp)
```

```python
with open('elect.csv','w') as file:
    file.write('{},{},{},'.format(driver.find_element_by_xpath('//*[@id="table01"]/thead/tr[1]/th[1]').text,
                                 driver.find_element_by_xpath('//*[@id="table01"]/thead/tr[1]/th[2]').text,
                                driver.find_element_by_xpath('//*[@id="table01"]/thead/tr[1]/th[3]').text))
    for f in people:
        file.write('{},'.format(f.text.replace('\n', ' ')))

    file.write('{},{},{}\n'.format(driver.find_element_by_xpath('//*[@id="table01"]/thead/tr[2]/th[14]').text,
        driver.find_element_by_xpath('//*[@id="table01"]/thead/tr[1]/th[5]').text.replace('\n',' '),
         driver.find_element_by_xpath('//*[@id="table01"]/thead/tr[1]/th[6]').text))
    for d in data:
        i=0
        for l in d:
            if (i != 0):
                file.write(',')
            file.write('{}'.format(l))
            i=1
        file.write('\n')
```
