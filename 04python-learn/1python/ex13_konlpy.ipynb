# 한글 형태소 분석 패키지 konlpy 사용해서 형태소 분리

```python
# 판다스로 저장한 csv파일 불러옴
import pandas as pd
youtube = pd.read_csv('c:/ai/workspace/python/youtube.csv', encoding='cp949')
```

```python
# 첫번째 게시글 제목 형태소 분석해보기
# 결과는 튜플의 리스트로 return
from konlpy.tag import Okt
twitter = Okt()
tagged = twitter.pos(youtube['title'][0], stem=True)
for i in range (0, len(tagged)):
    if (tagged[i][1] == 'Noun'):
        print(tagged[i][0])
```

```python
# seaborn 이용해서 데이터 시각화
import seaborn as sns
g = sns.catplot('view_num', data=youtube, kind='count', height=30)
```
