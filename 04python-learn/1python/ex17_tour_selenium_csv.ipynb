```python
import time
import re

isScore=re.compile("평점")
isComment=re.compile("상품평")

tac=int(driver.find_element_by_id("totalAllCnt").text)
count=0
if tac%10 == 0:
    count=tac//10
else:
    count=(tac//10)+1
    
myList=[]
for page in range(1,count+1):
    driver.execute_script("searchModule.SetCategoryList({},'')".format(page))
    time.sleep(2)
    items=driver.find_elements_by_css_selector(".panelZone>.oTravelBox>.boxList>li")
    for item in items:
        name=item.find_element_by_css_selector("h5.proTit").text.replace(","," ")
        price=int(item.find_element_by_css_selector("strong.proPrice").text.replace(" ","").split('원')[0].replace(",",""))
        print(item.find_element_by_css_selector("h5.proTit").text.replace(","," "))
        print(item.find_element_by_css_selector("strong.proPrice").text.replace(" ","").split('원')[0].replace(",",""))
        for i in item.find_elements_by_css_selector("p.proInfo"):
            if isScore.search(i.text):
                score=float(i.text.replace("평점 ",""))
                print("평점 : ",i.text.replace("평점 ",""))
            elif isComment.search(i.text):
                comment=int(i.text.replace("개의 상품평",""))
                print("상품평:",i.text.replace("개의 상품평",""))
        myList.append([name,price,score,comment])
        
myList
```

```python
import pandas
with open("shanghai.csv","w") as file:
    file.write("name,price,score,comment\n")
    for q in myList:
        file.write("{0},{1},{2},{3}\n".format(q[0],q[1],q[2],q[3]))
```
