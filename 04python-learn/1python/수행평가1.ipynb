```python
import requests
from bs4 import BeautifulSoup
response = requests.get('https://web.kma.go.kr/weather/observation/currentweather.jsp')
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table=soup.find(class_="table_develop3")
table
```

```python
with open('weather.csv','w') as file:
    file.write('지점,현재기온,체감온도,습도,풍향\n')
    for datas in table.find('tbody').find_all('tr'):
        data = datas.find_all('td')
        file.write(f'{data[0].text},{data[5].text},{data[7].text},{data[10].text},{data[11].text}\n')
```
