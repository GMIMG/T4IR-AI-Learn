```python
import numpy as np
import time
```

```python
X = np.array([1,2,3], dtype='float32')
X
```

```python
y = np.array([2,2.5, 3.5], dtype="float32")
y
```

```python
# 임의의 w, b 정함
w = 2
b = 1
learning_rate = 0.1
```

```python
hypothesis = w * X + b
hypothesis
```

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib import animation

# First set up the figure, the axis, and the plot element we want to animate
fig = plt.figure()
ax = plt.axes(xlim=(.5, 3.5), ylim=(0, 5))
# 예측데이터
line, = ax.plot([], [], lw=2)
# 원본데이터
line2, = ax.plot([], [], lw=2)

# initialization function: plot the background of each frame
def init():
    line.set_data([], [])
    return line, line2

# animation function.  This is called sequentially
def animate(i):
    global w, b, learning_rate
    hypothesis = w * X + b
    MSE = np.sum((hypothesis-y)**2)
    w_gred = 1/3*np.dot((hypothesis-y), X.T)
    b_gred = 1/3 * np.sum(hypothesis-y)
    w = w - (learning_rate*w_gred)
    b = b - (learning_rate*b_gred)
    line.set_data(X, hypothesis)
    line2.set_data(X, y)
    return line, line2

# call the animator.  blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, animate, init_func=init, frames=200, interval=20, blit=True)

anim.save('exAnimation.gif', writer='pillow', fps=30, dpi=100)

plt.show()
```
