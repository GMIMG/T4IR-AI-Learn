```python
import numpy as np
import tensorflow as tf
tf.enable_eager_execution()
```

```python
X = np.array([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]
],dtype = "float32")
```

```python
y = np.array([
    [0.],
    [0.],
    [0.],
    [1.]
], dtype="float32")
```

```python
w = np.array([2,3], dtype="float32")
b = np.array([4,5], dtype="float32")
```

```python
X*w+b
```

np.exp() ; 자연상수의 지수승

$np.exp(-h) == e^{-h}$

```python
def sigmoid(h):
    return 1/ (1 + np.exp(-h))
```

```python
hypothesis = sigmoid(X*w+b)
hypothesis
```

```python
hypothesis.T[0]
```

```python
X.T[0]
```

```python
w1_gred = 1/8 * np.sum(np.dot((hypothesis.T[0] - y.T), X.T[0]))
print("w_gred: ", w1_gred)
```

```python
w2_gred = 1/8 * np.sum(np.dot((hypothesis.T[1] - y.T), X.T[1]))
print("w_gred: ", w2_gred)
```

```python
b1_gred = 1/8 * np.sum(hypothesis[0]-y)
print('b1-gred: ', b1_gred)
```

```python
b2_gred = 1/8 * np.sum(hypothesis[1]-y)
print('b2_gred: ', b2_gred)
```

```python
learning_rate = 0.1

for i in range(100):
    for i in range(2):
        hypothesis = sigmoid(X*w+b)
        print('hypothesis: ', hypothesis)
        cost = -tf.reduce_mean(y * tf.log(1 - hypothesis) + (1 - y) * tf.log(1 - hypothesis))
        print('cost: ', cost)
        
        w_gred = 1/8 * np.sum(np.dot((hypothesis.T[i]-y.T), X.T[i]))
        print(f'w{i}_gred: {w_gred}')
        w[i] = w[i] - learning_rate * w_gred
        print(f'w{i} : {w}')
        b_gred = 1/8 * np.sum(hypothesis[i]-y)
        print(f'b{i}_gred: {b_gred}')
        b[i] = b[i] - learning_rate * b_gred
        print(f'b{i} : {b}')
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

tf.enable_eager_execution()
tf.__version__
```

```python
# 독립변수 X, 종속변수 Y
X = [1, 2, 3]
Y = [3, 5, 7]
```

```python
# tf.random_normal 로 난수생성
# tf.random_normal(생성할 난수의 형식(1개의 데이터가 저장된 배열), 평균, 표준편차)
W = tf.Variable(tf.random_normal([1], 0., 1.))
b = tf.Variable(tf.random_normal([1], 0., 1.))
```

```python
W.numpy()
```

```python
learning_rate = 0.1
```

```python
# 300번 반복해서 W, b 업데이트
for i in range(300):
    # with == (open, close)
    with tf.GradientTape() as tape:
        hypothesis = W * X + b
        print("hypothesis:",hypothesis)
        cost = tf.reduce_mean(tf.square(hypothesis - Y))
        print('cost:',cost)
    W_grad, b_grad = tape.gradient(cost, [W, b])
    print('W_grad:', W_grad, ":b_grad:", b_grad)
    # W,b 값을 감소
    W.assign_sub(W_grad*learning_rate)
    b.assign_sub(b_grad*learning_rate)
```
