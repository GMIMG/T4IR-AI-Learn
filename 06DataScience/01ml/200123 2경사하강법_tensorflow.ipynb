## tensorflow 사용

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

tf.enable_eager_execution()
tf.__version__
```

```python
# 독립변수 X, 종속변수 Y
X = [1, 2, 3]
Y = [2,2.5, 3.5]
```

```python
# tf.random_normal 로 난수생성
# tf.random_normal(생성할 난수의 형식(1개의 데이터가 저장된 배열), 평균, 표준편차)
W = tf.Variable(tf.random_normal([1], 0., 1.))
b = tf.Variable(tf.random_normal([1], 0., 1.))
```

```python
W.numpy()
```

```python
learning_rate = 0.1
```

```python
# 300번 반복해서 W, b 업데이트
for i in range(300):
    # with == (open, close)
    with tf.GradientTape() as tape:
        hypothesis = W * X + b
        print("hypothesis:",hypothesis)
        cost = tf.reduce_mean(tf.square(hypothesis - Y))
        print('cost:',cost)
    W_grad, b_grad = tape.gradient(cost, [W, b])
    print('W_grad:', W_grad, ":b_grad:", b_grad)
    # W,b 값을 감소
    W.assign_sub(W_grad*learning_rate)
    b.assign_sub(b_grad*learning_rate)
    print('W: ',W)
    print('b: ',b)
    print('='*50)
```
