```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from tensorflow.nn import relu
import tensorflow as tf
tf.enable_eager_execution()
```

```python
data = datasets.load_breast_cancer()
data
```

```python
X = np.array(data.data, dtype='float32')
X
```

```python
y = np.array(data.target, dtype='float32')
```

```python
y = y.reshape(-1,1)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
def sigmoid(h):
    return 1 / (1 + np.exp(-h))
```

```python
# w와 b를 초기화 시켜주는 객체
initializer = tf.contrib.layers.xavier_initializer()

w = tf.Variable(initializer([30,1]))
w
```

```python
b = tf.Variable(initializer([1]))
b
```

```python
tf.matmul(X_train, w)+b
```

```python
hypothesis = tf.sigmoid(tf.matmul(X_train, w) + b)
hypothesis
```

```python
optimizer = tf.train.AdamOptimizer(0.001)
```

```python
for step in range(100):
    with tf.GradientTape() as tape:
        hypothesis = tf.sigmoid(tf.matmul(X_train, w) + b)
        cost = -tf.reduce_mean(y_train * tf.log(hypothesis) + (1 - y_train) * tf.log(1 - hypothesis))
        grads = tape.gradient(cost, [w,b])
    optimizer.apply_gradients(grads_and_vars = zip(grads, [w,b]))
    
    if step % 10 == 0:
        print("=" * 50)
        print(f"step: {step}, cost : {cost}, w : {w.numpy()}, b : {b.numpy()}")
        print('=' * 50)
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
predict = tf. sigmoid(tf.matmul(X_test, w) + b)
predict
```

```python
predict01  =tf.cast(predict >0.5, dtype = tf.float32)
ac01 = tf.equal(predict01, y_test)
ac01
```

```python
ac02 = tf.cast(ac01, dtype='float32')
ac03 = tf.reduce_mean(ac02)
ac03
```
