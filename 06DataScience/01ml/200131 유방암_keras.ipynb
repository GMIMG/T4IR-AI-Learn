```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
import tensorflow as tf
```

```python
data = datasets.load_breast_cancer()
data
```

```python
X = data.data
y = data.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
model = Sequential()
model.add(Dense(1,input_dim=30, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.001),metrics=['acc'])
```

```python
model.fit(X_train, y_train, epochs=1000)
```

```python
X_test = scaler.transform(X_test)
```

```python
pred = model.predict(X_test)
pred
```

```python
predict01 = np.where(pred>0.5, 1, 0)
print('=========================')
print('predict01')
print(predict01)
```

```python
predict02 = predict01.flatten()
predict02
```

```python
predict03 = (predict02 == y_test)
predict03
```
