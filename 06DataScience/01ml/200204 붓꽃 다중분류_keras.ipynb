```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
data = datasets.load_iris()
data
```

```python
X = data.data
```

```python
y = data.target
```

```python
y = y.reshape(-1, 1)
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
X_train
```

```python
from sklearn.preprocessing import OneHotEncoder
enc = OneHotEncoder(categories = 'auto')
enc.fit(y)
```

```python
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot
```

```python
y_train_onehot = np.array(y_train_onehot, dtype='float32')
```

```python
y_train_onehot
```

```python
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot = np.array(y_test_onehot, dtype='float32')
```

```python
model = Sequential()

model.add(Dense(16, input_dim = 4, activation = 'sigmoid'))
model.add(Dense(3,activation = 'softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer=Adam(lr=0.1),metrics=['acc'])
```

```python
model.fit(X_train, y_train_onehot, epochs = 1000)
```

```python
X_test = scaler.transform(X_test)
```

```python
pred = model.predict(X_test)
```

```python
np.where(pred>0.5, 1, 0)
```

```python
predict = np.argmax(pred,1)
```

```python
target = np.argmax(y_test_onehot, 1)
```

```python
predict02 = (predict == target)
```
